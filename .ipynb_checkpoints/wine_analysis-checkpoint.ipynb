## 2. Cosine similarity

➤ Cosine similarity — measures the similarity between two samples based on the angle between their feature vectors. Values range from -1 (opposite) to 1 (identical), helping to find wines with similar chemical profiles.



Firstly replace the only nin-numeric variable with dummy- variables

```python
wine = pd.get_dummies(wine, columns = ['wine_type'], drop_first = True)
wine
```

```python
X = wine.drop('quality', axis = 1)
X.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pd.DataFrame(X_scaled)
```

```python
similarity_matrix = cosine_similarity(X_scaled)
sim_df = pd.DataFrame(similarity_matrix)
sim_df.iloc[:5,:5]
```

### Recommend 5 wines most similar to the given one

```python
def find_similar_indexes(similarity_matrix, index):
    return similarity_matrix.iloc[index].sort_values(ascending = False).index[:5]
```

Example for wine with id *5* :

```python
id = 5
wine.iloc[find_similar_indexes(sim_df, id)]
```

### Function, that finds similar wines by index and entered quality

```python
def recommend_wines_by_quality(df, similarity_matrix, target_quality, index):
    subset_ids = df[df['quality'] == target_quality].index
    sim_wines = similarity_matrix.iloc[index]
    sim_filtered_wines = sim_wines[subset_ids].sort_values(ascending = False).index
    return sim_filtered_wines
    
    
```

*Example* :

5  Wines, that are similar to wine with id 8 and that have quality 8

```python
id = 8
target_quality = 8
n = 5

pd.options.display.max_columns = None
pd.options.display.max_rows = None

print(f"Similar wines with quality {target_quality}:\n")
wine.iloc[recommend_wines_by_quality(wine, sim_df, target_quality, id)][:n]
```

### Function, that compares two wines on all features

```python
def compare_wines(similarity_matrix, id1, id2):
    return round(similarity_matrix.iloc[id1, id2], 3)
```

Example of work:

```python
id1 = 0
id2 = 999
print(f"Wines with indexes {id1} and {id2} are {compare_wines(sim_df, id1, id2) * 100} % similar")
```
